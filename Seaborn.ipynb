```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.distplot(iris.sepal_length)
plt.show()
```

```python
sns.regplot(iris.sepal_length, iris.petal_length)
plt.title("A Title") #Add plot title
plt.ylabel("y-label") #Adjust the label of the y-axis
plt.xlabel("x-label") #Adjust the label of the x-axis
plt.ylim(0,10) #Adjust the limits of the y-axis
plt.xlim(0,10) #Adjust the limits of the x-axis
sns.set_style('ticks')
plt.show()
```

```python
sns.pairplot(iris, kind="reg")
```

```python
sns.pairplot(iris, x_vars = ['sepal_length'], y_vars=['petal_length'], size = 5, kind="reg")
plt.show()
```

```python
sns.pairplot(iris,  x_vars = ['sepal_length'], y_vars=['petal_length'], size = 5,  hue='species')
plt.show()
```

```python
sns.pairplot(iris, x_vars=['sepal_length'], y_vars = ['petal_length'], size=5, hue='species')
plt.title('The Iris data set: petal length vs sepal length')
plt.show()
```

```python
sns.pairplot(iris, x_vars=['sepal_length'], y_vars = ['petal_length'], size=5, hue='species', palette = 'pastel' )
```

```python
grid = sns.FacetGrid(iris, col='species')
grid = grid.map(plt.scatter, "sepal_length", "petal_length")
plt.show()
```

```python
sns.violinplot(x = iris.species, y=iris.petal_length)
```

```python
sns.swarmplot(x = "species", y="sepal_length", data=iris)
```

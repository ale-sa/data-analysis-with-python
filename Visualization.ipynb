```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi)
Y = np.sin(X) + np.cos(X)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
plt.plot(X,Y, color='red', linewidth=2.0)
plt.xlim(-4,4)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.show()
```

```python
plt.plot(X,Y, color='red', linewidth=2.0)
plt.xlim(-4,4)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'])
plt.show()
```

```python
plt.plot(X,Y, color='red', linewidth=2.0)
plt.xlim(-4,4)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'])
plt.title('The function $f(x)=\sin(x) + \cos(x)$')
plt.show()
```

```python
foo = np.linspace(-15,15)
```

```python
line1 = np.sin(foo)/foo
```

```python
line2 = line1 * 2
```

```python
line3 = line1 * 3
```

```python
plt.plot(foo, line1, color="blue", linestyle="dashed", label="$\sin(x)/x$")
plt.plot(foo, line2, color="green", label="$2 * \sin(x)/x$")
plt.plot(foo, line3, color="red", marker="o", label="$3 * \sin(x)/x$")
plt.ylim(-1,3.5)
plt.yticks([-1,0,1,2,3])
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(foo, line1, color="blue", linestyle="dashed", label="$\sin(x)/x$")
ax.set_title("Test")
plt.show()
```

```python
countries = ['USA', 'GB', 'China', 'Russia', 'Germany', 'France']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 17, 10])
golds = np.array([46, 27, 26, 19, 17])
total = np.array([121,67,70,55,42])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Section A
# import data
data = pd.DataFrame.from_csv("data.csv")
```

```python
# Column names
data.columns
```

```python
# Data Types
data.dtypes
```

```python
# Summarize data
data.describe()
```

```python
# Check for NAs
nas = data.isnull()
data.isnull().sum()
```

```python
# Remove NAs
data.dropna(how="any",axis=0, inplace=True)
# Check for NAs again
data.isnull().sum()
```

```python
# Data is already sorted by happiness rank, so first 10 are happiest countries
data.head(10)
```

```python
# Ten least happy countries
data.tail(10)
```

```python
## Part B
# Subset to specific regions
by_region = data.loc[data['Region'].isin(["North America","Latin America", "Western Europe", "Eastern Europe", "Asia Pacific", "Africa"])]
```

```python
# Get mean happiness score by region and sort it
meansByRegion = by_region.groupby("Region")["Happiness Score"].mean()
meansByRegion.sort_values(ascending=False)
```

```python
# Countries with score over 6
happyCountries = by_region[by_region["Happiness Score"] > 6]
# Count countries with score over 6
happyCountries.groupby("Region")["Happiness Rank"].agg("count")
```

```python
# Get min and max by region
max = by_region.groupby("Region")["Happiness Score"].max()
min = by_region.groupby("Region")["Happiness Score"].min()
```

```python
# Africa has the widest range of happiness scores
max - min
```

```python
## Part C - Bar plot of happiness score
# get top ten rows
topTen = data.ix[:10,]
index = np.arange(10)
```

```python
# Create horizontal Bar Chart
plt.barh(index, topTen["Happiness Score"])
plt.yticks(index, topTen.index)
plt.title("Happiness Scores for Top Ten Happiest Countries")
plt.xlim(0,8)
plt.xlabel("Happiness Score")
plt.ylabel("Country")
```

```python
# Stacked Bar Chart of top ten happiest countries
plt.figure(figsize=(10,5))
plt.barh(index, topTen["Economy"], color="green", label="Economy")
plt.barh(index, topTen["Family"], color="orange", label="Family", left=topTen["Economy"])
plt.barh(index, topTen["Health"], color="red", label="Health", left=topTen["Economy"] + topTen["Family"])
plt.barh(index, topTen["Freedom"], color="purple", label="Freedom", left=topTen["Economy"] + topTen["Family"] + topTen["Health"])
plt.barh(index, topTen["Generosity"], color="blue", label="Generosity", left=topTen["Economy"] + topTen["Family"] + topTen["Health"] + topTen["Freedom"])
plt.barh(index, topTen["Corruption"], color="yellow", label="Corruption", left=topTen["Economy"] + topTen["Family"] + topTen["Health"] + topTen["Freedom"] + topTen["Generosity"])
plt.barh(index, topTen["Dystopia"], color="brown", label="Dystopia", left=topTen["Economy"] + topTen["Family"] + topTen["Health"] + topTen["Freedom"] + topTen["Generosity"] + topTen["Corruption"])
plt.legend(loc="best")
plt.yticks(index, topTen.index)
plt.xlim(0,8)
plt.ylabel("Country")
plt.xlabel("Happiness Score")
plt.title("Happiness for Top Ten Countries by Feature")
plt.show()
```
